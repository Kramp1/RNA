```python
import numpy as np
from numpy.linalg import pinv
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
class ELM:

    def __init__(self, input_dim, hidden_dim, output_dim):
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.W = np.random.randn(self.input_dim, self.hidden_dim)
        self.b = np.random.randn(self.hidden_dim)
        self.beta = None
        self.activation = lambda x: 1 / (1 + np.exp(-x))
    
    def fit(self, X, y):
        H = self.activation(np.dot(X, self.W) + self.b)
        self.beta = np.dot(pinv(H), y)

    def predict(self, X):
        H = self.activation(np.dot(X, self.W) + self.b)
        output = np.dot(H, self.beta)
        print("Output shape:", output.shape)
        return (output > 0.5).astype(int)
```

```python
# Plotting
def plot_decision_boundary(blobELM1, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid_points = np.c_[xx.ravel(), yy.ravel()]

    Z = blobELM1.predict(grid_points)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

    plt.scatter(X[y.flatten() == 0][:, 0], X[y.flatten() == 0][:, 1], color='blue', label="Class 1")
    plt.scatter(X[y.flatten() == 1][:, 0], X[y.flatten() == 1][:, 1], color='red', label="Class 2")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
#mlbench.2dnormals(200)
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, centers=2, n_features=2, random_state=42)
```

```python
#2 neur√¥nios

blobELM1 = ELM(2, 2, 1)
blobELM1.fit(X, y)
plot_decision_boundary(blobELM1, X, y, "Dataset 1")
```

```python
#mlbench.spirals(100,sd = 0.05)
import numpy as np

def make_spirals(n_samples, noise=0.05):
    n = np.sqrt(np.random.rand(n_samples // 2)) * 780 * (2 * np.pi) / 360
    d1x = -np.cos(n) * n + np.random.rand(n_samples // 2) * noise
    d1y = np.sin(n) * n + np.random.rand(n_samples // 2) * noise
    x = np.vstack((np.append(d1x, -d1x), np.append(d1y, -d1y))).T
    y = np.hstack((np.zeros(n_samples // 2), np.ones(n_samples // 2)))
    return x, y

X, y = make_spirals(100, noise=0.05)
```

```python
#5 neur√¥nios
spiralELM1 = ELM(2, 5, 1)
spiralELM1.fit(X, y)
plot_decision_boundary(spiralELM1, X, y, "Dataset 3")
```
